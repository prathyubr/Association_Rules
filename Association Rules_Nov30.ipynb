### Get Best Associates of France

```python
france_data = retail_data[retail_data['Country'] == 'France']
france_data
```

```python
france_data_pivot_table = pd.pivot_table(data = france_data,values='Quantity',index='InvoiceNo',columns='Description').fillna(0)
france_data_pivot_table
```

```python
france_encoded_data = france_data_pivot_table.applymap(lambda x:1 if x>0 else 0)
france_encoded_data
```

## 4. Data Preparation

```python
pd.pivot_table(data = uk_data,values='Quantity',index='InvoiceNo',columns='Description').fillna(0)
```

## 5. Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_itemsets = apriori(df =  france_encoded_data,min_support=0.02,use_colnames=True) #apriori(df,min_support=0.5,)
freq_itemsets
```

```python
freq_itemsets.to_csv('Frequent_itemsets.csv')
```

```python
best_assosiates = association_rules(df = freq_itemsets,metric='confidence',min_threshold=0.70,)
best_assosiates
```

```python
best_assosiates.to_csv('best_associates_france.csv')
```
